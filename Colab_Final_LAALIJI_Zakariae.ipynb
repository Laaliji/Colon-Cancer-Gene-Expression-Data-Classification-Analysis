```python
# Function to count outliers using IQR method
def count_outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = series[(series < lower_bound) | (series > upper_bound)].count()
    return outliers

# Calculate outliers for each gene
outlier_counts = features.apply(count_outliers)

# Display results
print("Number of Outliers per Gene:")
print(outlier_counts)
```

Outlier counts vary across the 60 genes, with some showing significant outliers (e.g., TOR4A: 76, ESRP1: 73, BSPRY: 68, EPS8L1: 41) and others none (e.g., ADH1C, SLC7A5, RNF43). High outlier counts (e.g., 9% of samples for TOR4A) suggest potential preprocessing (clipping or robust scaling) to reduce their impact on model performance, while genes with fewer or no outliers can be modeled directly.

### **Data Splitting and Scaling**

**Reasons for Standardization :**

1.   **Scale-Sensitive Models:** Algorithms like Support Vector Machines (SVM), k-Nearest Neighbors (k-NN), and neural networks assume features are on the same scale. Without standardization, genes with larger ranges (e.g., DAO: 1.16–10.08) would dominate over those with smaller ranges (e.g., PLLP: 0.12–4.76), skewing distance-based calculations and weight assignments.
2.   **Improved Convergence:** Gradient-based optimization (e.g. logistic regression) converges faster when features are standardized. Unscaled data with high variance (e.g., ADH1C std: 2.89) can lead to slow or unstable learning.
3. **Outlier Handling:** Standardization reduces the impact of outliers by normalizing their relative position (e.g., an outlier in TOR4A becomes a z-score, typically within ±3). This helps models focus on patterns rather than extreme values, though robust scaling (e.g., RobustScaler) can be considered if outliers persist.

#### Logistic Regression

Logistic Regression is a statistical method used for binary classification, predicting the probability that a sample belongs to a particular class (e.g., `tumoral` vs. `normal`). Unlike linear regression, it models the relationship between features and the log-odds of the outcome, making it suitable for classification tasks.

**Theory:**

Logistic Regression assumes a linear relationship between the features and the log-odds of the target variable. It uses the sigmoid function to transform this linear combination into a probability between 0 and 1. The model then applies a threshold (typically 0.5) to classify samples into one of two classes.

**Mathematics**:

Given a feature vector $\mathbf{x} = [x_1, x_2, \ldots, x_n]$ and weights $\mathbf{w} = [w_1, w_2, \ldots, w_n]$, plus a bias term $b$, the linear combination is:

$$z = \mathbf{w}^T \mathbf{x} + b = w_1 x_1 + w_2 x_2 + \cdots + w_n x_n + b$$

The probability of the positive class (`tumoral`) is computed using the sigmoid function:

$$P(y = 1 | \mathbf{x}) = \sigma(z) = \frac{1}{1 + e^{-z}}$$

The probability of the negative class (`normal`) is:

$$P(y = 0 | \mathbf{x}) = 1 - P(y = 1 | \mathbf{x})$$

The model is trained by minimizing the log-loss (or cross-entropy loss):

$$L = -\frac{1}{m} \sum_{i=1}^m \left[ y_i \log(\hat{y}_i) + (1 - y_i) \log(1 - \hat{y}_i) \right]$$

where $m$ is the number of samples, $y_i$ is the true label, and $\hat{y}_i = P(y_i = 1 | \mathbf{x}_i)$.

**Coefficient Analysis with Stability**:

The coefficients (\( w_j \)) in Logistic Regression indicate the influence of each gene on the log-odds of the `tumoral` class. To address potential overfitting observed with all 60 genes (e.g., accuracy of 1.00), we perform a dynamic feature extraction process. Instead of selecting a fixed number of genes (e.g., 10), we incrementally add genes ranked by absolute coefficient values and evaluate the model’s accuracy on the test set. This approach identifies the point where accuracy stabilizes, ensuring a balance between model performance and generalizability while reducing dimensionality.

**Methodology**:

- Rank genes by the absolute value of their coefficients.
- Iteratively train the model on increasing subsets of top-ranked genes (e.g., 1, 2, 3, ..., up to 60).
- Monitor accuracy to detect a plateau, indicating the optimal number of features.
- This method leverages the standardized data to focus on the most impactful genes, enhancing interpretability and robustness for colon cancer classification.

**Implementation**:

The analysis uses the trained Logistic Regression model’s coefficients to rank genes, followed by iterative retraining and evaluation to determine the stability point.

#### SVM (Support Vector Machine)

* **Theory :** Support Vector Machine (SVM) is a supervised learning algorithm used for binary classification. It aims to find the optimal hyperplane that best separates two classes (e.g., tumoral vs. normal) by maximizing the margin between the classes. The margin is defined as the distance between the hyperplane and the nearest data points from each class, called support vectors.

* **Key Concept**: SVM seeks to maximize the margin while minimizing classification errors. For non-linearly separable data, it uses the "soft margin" approach (allowing some misclassifications) or a kernel trick (e.g., RBF kernel) to transform the data into a higher-dimensional space where a linear boundary exists.

* **Linear SVM**: Since we aim to extract feature importance (gene influence), we'll use a linear kernel, where the decision boundary is a hyperplane directly interpretable via coefficients.

* **Mathematics**:
   * Given a feature vector $\mathbf{x} = [x_1, x_2, \ldots, x_n]$, weights $\mathbf{w} = [w_1, w_2, \ldots, w_n]$, and bias $b$, the decision function is: $f(\mathbf{x}) = \mathbf{w}^T \mathbf{x} + b$
   
   * A sample is classified as:
     \begin{cases}
     \text{tumoral} & \text{if } f(\mathbf{x}) \geq 0 \\
     \text{normal} & \text{if } f(\mathbf{x}) < 0
     \end{cases}
     
   * SVM optimizes the following objective (soft margin):
     $$\min_{\mathbf{w}, b, \xi} \frac{1}{2} \|\mathbf{w}\|^2 + C \sum_{i=1}^m \xi_i$$
     
     subject to:
     $$y_i (\mathbf{w}^T \mathbf{x}_i + b) \geq 1 - \xi_i, \quad \xi_i \geq 0, \quad \forall i$$
     
     where $C$ is the regularization parameter (trade-off between margin maximization and classification error), $\xi_i$ are slack variables, and $y_i \in \{-1, 1\}$ are the labels (normal = -1, tumoral = 1).

#### K-Nearest Neighbors (KNN)

**Theory**

KNN is a lazy learning algorithm, meaning it does not explicitly build a model during the training phase. Instead, it stores the training data and makes predictions by finding the kk nearest data points (neighbors) to a new, unseen sample based on a distance metric. The class of the new sample is then determined by a majority vote among its kk neighbors. This approach is particularly useful for datasets with complex decision boundaries or when the underlying distribution is unknown, as is often the case with high-dimensional biological data like gene expressions.

**Mathematics**

**Distance Metrics**

The core of KNN lies in measuring the distance between data points. The most common metric is the Euclidean distance, defined as:

$$d(x,x')=\sqrt{\sum_{i=1}^{n}(x_i-x'_i)^2}$$

where:
- $x$ and $x'$ are two data points (e.g., gene expression vectors with 60 dimensions).
- $x_i$ and $x'_i$ are the values of the $i$-th feature (gene) for $x$ and $x'$.
- $n$ is the number of features (60 in my case).

Other metrics, such as Manhattan distance:

$$d(x,x')=\sum_{i=1}^{n}|x_i-x'_i|$$

or Minkowski distance (a generalization):

$$d(x,x')=\left(\sum_{i=1}^{n}|x_i-x'_i|^p\right)^{1/p}$$

(where $p=2$ gives Euclidean distance, $p=1$ gives Manhattan), can also be used. The choice depends on the data's geometry, with Euclidean being standard for gene expression unless specific patterns suggest otherwise.

**Prediction Process**

For a test sample $x_{test}$:

1. Compute the distance from $x_{test}$ to all points in the training set $\{x_1,x_2,...,x_m\}$, where $m$ is the number of training samples.
2. Select the $k$ nearest neighbors based on the smallest distances.
3. Assign the class label by majority vote:

$$\text{Class}(x_{test})=\arg\max_{c \in C}\sum_{i \in N_k}I(y_i=c)$$

where:
- $C$ is the set of possible classes ('tumoral', 'normal').
- $N_k$ is the set of $k$ nearest neighbors.
- $y_i$ is the class label of the $i$-th neighbor.
- $I$ is the indicator function (1 if true, 0 if false).

4. Optionally, weighted voting can be used, where closer neighbors contribute more:

$$\text{Class}(x_{test})=\arg\max_{c \in C}\sum_{i \in N_k}w_i I(y_i=c)$$

where $w_i=1/d(x_{test},x_i)$ or $w_i=1/d(x_{test},x_i)^2$ (inverse distance or squared inverse distance).

**Computational Complexity**

- Training: $O(1)$, as KNN simply stores the training data.
- Prediction: $O(m \cdot n)$, where $m$ is the number of training samples and $n$ is the number of features (60). For large $m$ (e.g., 804 samples), this can be optimized using data structures like KD-trees or Ball trees, reducing complexity to $O(\log m \cdot n)$ in lower dimensions.

**Choosing $k$**

- $k$ is typically chosen via cross-validation. A small $k$ (e.g., 1 or 3) makes the model sensitive to noise, while a large $k$ (e.g., 15 or 20) smooths the decision boundary but may include irrelevant neighbors. For my dataset, $k$ should be tuned to balance noise sensitivity and generalization, likely tested in the range of 1 to 20.

**Handling High-Dimensional Data**

- With 60 genes, the curse of dimensionality may dilute distance metrics, as all points become equidistant in high dimensions. Feature selection (e.g., using the optimal genes from SVM/Logistic Regression) or dimensionality reduction (e.g., PCA) can mitigate this, focusing KNN on the most discriminative features.

```python
# Rank genes using Logistic Regression coefficients
log_model = LogisticRegression(random_state=42)
log_model.fit(X_train_scaled, y_train)
log_coefficients = pd.Series(log_model.coef_[0], index=X_train_scaled.columns)
log_abs_coefficients = log_coefficients.abs()
log_sorted_genes = log_abs_coefficients.sort_values(ascending=False).index.tolist()

# Test KNN with different numbers of features
feature_counts = [5, 10, 20, 30, 40, 50, 60]
cv_accuracies_dim = []
best_k_values = []

print("=== KNN: Curse of Dimensionality Analysis ===")
for n in feature_counts:
    selected_genes = log_sorted_genes[:n]
    X_train_subset = X_train_scaled[selected_genes]
    k_values = range(1, 21)
    cv_scores_k = []
    for k in k_values:
        knn = KNeighborsClassifier(n_neighbors=k, metric='euclidean')
        cv_scores = cross_val_score(knn, X_train_subset, y_train, cv=5, scoring='accuracy')
        cv_scores_k.append(cv_scores.mean())
    best_k = k_values[np.argmax(cv_scores_k)]
    best_cv = max(cv_scores_k)
    cv_accuracies_dim.append(best_cv)
    best_k_values.append(best_k)
    print(f"Features={n:<2} | Best k={best_k:<2} | Best CV Accuracy: {round(best_cv, 4)}")
```

#### Decision Tree

**Mathematical Theory and Framework of Decision Trees**

*1. Overview of Decision Trees*

Decision Trees are a non-parametric supervised learning method used for classification and regression tasks. In classification, they recursively partition the feature space into regions based on feature values, creating a tree-like structure where each path from the root to a leaf represents a decision rule leading to a class label. For my binary classification task (tumoral vs. normal), the tree will assign each sample to one of two classes by traversing decision rules based on gene expression values.

* **Structure:**
    * Root Node: Represents the entire dataset (all 60 genes and samples).
    * Internal Nodes: Represent decision points, splitting the data based on a feature and threshold (e.g., $Gene1 \leq 0.5$).
    * Leaf Nodes: Represent the final class label (e.g., tumoral or normal).
    * Edges: Connect nodes, representing the outcome of a decision rule (e.g., true or false).

*2. Splitting Criteria: Measuring Impurity*

The core of a Decision Tree lies in how it chooses to split the data at each node. Splits are made to maximize the homogeneity (purity) of the resulting child nodes, measured using an impurity function. The most common impurity measures for classification are:

* **Gini Impurity:**
    * Measures the probability of incorrectly classifying a sample if it were randomly labeled according to the node's class distribution.
    * Formula for a node $t$ with $K$ classes (here, $K=2$, tumoral and normal):
    $$Gini(t)=1-\sum_{k=1}^{K}p_k^2$$
    where $p_k$ is the proportion of class $k$ in node $t$.
    * Example: If a node has 50 tumoral and 50 normal samples ($p_{tumoral}=0.5$, $p_{normal}=0.5$):
    $$Gini(t)=1-(0.5^2+0.5^2)=1-(0.25+0.25)=0.5$$
    * Range: 0 (pure node, all samples of one class) to 0.5 (maximum impurity for two classes).

* **Entropy (Information Gain):**
    * Measures the uncertainty or randomness in a node, used to compute Information Gain.
    * Formula for a node $t$:
    $$Entropy(t)=-\sum_{k=1}^{K}p_k\log_2(p_k)$$
    * Using the same example (50% tumoral, 50% normal):
    $$Entropy(t)=-(0.5\log_2(0.5)+0.5\log_2(0.5))=-(0.5\times(-1)+0.5\times(-1))=1.0$$
    * Range: 0 (pure node) to 1 (maximum uncertainty for two classes).
    * Information Gain: The reduction in entropy after a split:
    $$Information\ Gain=Entropy(parent)-\sum_{i\in\{left,\ right\}}\frac{|D_i|}{|D|}Entropy(D_i)$$
    where $|D_i|$ and $|D|$ are the sizes of the child and parent datasets, respectively.

* **Choice of Criterion:** Gini impurity is computationally faster (no logarithms) and often yields similar results to entropy for binary classification. For my dataset, we'll use Gini (default in scikit-learn), but we can test entropy later if needed.

*3. Splitting Process*

At each node, the Decision Tree evaluates all features (60 genes) and possible thresholds (e.g., $Gene1 \leq 0.1$, $Gene1 \leq 0.2$, etc.) to find the split that minimizes the weighted impurity of the child nodes:

$$Weighted\ Impurity=\frac{|D_{left}|}{|D|}Impurity(D_{left})+\frac{|D_{right}|}{|D|}Impurity(D_{right})$$

* The algorithm selects the feature and threshold yielding the lowest weighted impurity.
* Example: If splitting on $Gene1 \leq 0.5$ reduces the weighted Gini from 0.5 to 0.2, while $Gene2 \leq 1.0$ reduces it to 0.3, the tree chooses $Gene1$.

*4. Stopping Criteria*

The tree stops growing when one of the following conditions is met:

* All samples in a node belong to the same class (pure node).
* A predefined maximum depth ($max\_depth$) is reached (e.g., 5, to prevent overfitting).
* The number of samples in a node falls below a minimum threshold ($min\_samples\_split$, e.g., 2).
* The number of samples required to form a leaf node is below a threshold ($min\_samples\_leaf$).
* Without constraints, the tree grows until all leaves are pure, risking overfitting by memorizing noise in the training data.

*5. Prediction*

To classify a new sample:

* Start at the root node.
* Traverse the tree by following decision rules (e.g., if $Gene1 \leq 0.5$, go left; else, go right).
* Reach a leaf node and assign its majority class (e.g., tumoral).
* For my dataset, a rule might be: If $Gene1 \leq 0.5$ and $Gene2 > 1.0$, classify as tumoral; else, normal.

*6. Feature Importance for Extraction*

Decision Trees are particularly useful for feature extraction because they naturally rank features based on their contribution to reducing impurity:

* **Feature Importance Formula:**
$$I(f)=\sum_{nodes\ using\ f}\frac{N_t}{N}\Delta Impurity(t)$$
    * $I(f)$: Importance of feature $f$.
    * $N_t$: Number of samples at node $t$.
    * $N$: Total number of samples.
    * $\Delta Impurity(t)$: Impurity reduction (e.g., Gini) at node $t$ due to the split on $f$.
* **Normalization:** Importance scores are normalized to sum to 1 across all features.
* **Interpretation:**
    * Features used higher in the tree (near the root) or frequently across nodes have higher importance, as they affect more samples and reduce impurity more significantly.
    * For my 60-gene dataset, a gene like $Gene1$ with a high importance score (e.g., 0.3) indicates it plays a major role in separating tumoral and normal samples.

#### Random Forest

**Mathematical Theory and Framework of Random Forests**

*1. Overview of Random Forests*

Random Forest is an ensemble learning method that constructs multiple decision trees during training and outputs the class that is the mode of the classes (for classification) or mean prediction (for regression) of the individual trees. Random Forests overcome the overfitting problem of individual decision trees by introducing two sources of randomness: bootstrap sampling of training data and random feature selection at each node.

*2. Mathematical Framework*

* **Ensemble Construction:**
  * Given a training set $X = \{x_1, x_2, ..., x_n\}$ with labels $Y = \{y_1, y_2, ..., y_n\}$
  * For $b = 1$ to $B$ (number of trees):
    * Draw a bootstrap sample $X_b$ of size $n$ by sampling with replacement from $X$
    * Grow a decision tree $T_b$ to the bootstrapped data by recursively repeating the following steps for each terminal node until the minimum node size is reached:
      * Select $m$ features at random from the $p$ available features, where typically $m \approx \sqrt{p}$ for classification
      * Pick the best feature/split-point among the $m$ features
      * Split the node into two daughter nodes

* **Bootstrap Aggregating (Bagging):**
  * Each tree is trained on a bootstrap sample $X_b$ drawn with replacement from the training set
  * Approximately $\frac{1}{3}$ of the samples (called Out-Of-Bag or OOB samples) are not used for training each tree
  * If $N$ is the size of the training set, the probability that a sample is not selected in a bootstrap sample is:
  $$P(\text{not selected}) = \left(1 - \frac{1}{N}\right)^N \approx e^{-1} \approx 0.368$$
  * The expected fraction of unique samples in each bootstrap sample is approximately $1 - e^{-1} \approx 0.632$

* **Random Feature Selection:**
  * At each node, only a random subset of features is considered for splitting
  * For classification with $p$ features, the common choice is $m = \sqrt{p}$ or $m = \log_2(p)$
  * This feature randomization decorrelates the trees, reducing the variance of the ensemble

*3. Prediction*

* **Classification:**
  * The prediction for a new sample $x$ is the majority vote of the predictions from all trees:
  $$\hat{y} = \text{mode}\{T_b(x) : b = 1, 2, ..., B\}$$
  * Class probabilities can be calculated as the proportion of trees voting for each class:
  $$P(y = j | x) = \frac{1}{B}\sum_{b=1}^{B}I(T_b(x) = j)$$
  where $I(\cdot)$ is the indicator function.

* **Regression:**
  * The prediction is the average of predictions from all trees:
  $$\hat{y} = \frac{1}{B}\sum_{b=1}^{B}T_b(x)$$

*4. Out-Of-Bag (OOB) Error Estimation*

* For each sample $(x_i, y_i)$ in the training set, consider only the trees $T_b$ where $x_i$ was not in the bootstrap sample (approximately $\frac{B}{3}$ trees)
* The OOB prediction for $x_i$ is:
  * For classification: $\hat{y}_i^{OOB} = \text{mode}\{T_b(x_i) : x_i \notin X_b\}$
  * For regression: $\hat{y}_i^{OOB} = \frac{1}{|\{b : x_i \notin X_b\}|}\sum_{b : x_i \notin X_b}T_b(x_i)$
* The OOB error estimate is:
  * For classification: $\text{OOB Error} = \frac{1}{n}\sum_{i=1}^{n}I(\hat{y}_i^{OOB} \neq y_i)$
  * For regression: $\text{OOB Error} = \frac{1}{n}\sum_{i=1}^{n}(\hat{y}_i^{OOB} - y_i)^2$
* This provides an unbiased estimate of the generalization error without requiring a separate validation set

*5. Feature Importance*

Random Forests provide two main methods for calculating feature importance:

* **Mean Decrease in Impurity (MDI):**
  * For each feature $j$, the importance is calculated as:
  $$I(j) = \frac{1}{B}\sum_{b=1}^{B}\sum_{t \in T_b}\Delta i(s_t, j)$$
  where:
    * $\Delta i(s_t, j)$ is the decrease in impurity (Gini or entropy) at node $t$ due to a split $s_t$ on feature $j$
    * The sum is over all nodes $t$ in tree $T_b$ where feature $j$ is used for splitting
  * This importance is normalized to sum to 1 across all features

* **Mean Decrease in Accuracy (MDA) or Permutation Importance:**
  * For each tree $T_b$, calculate the OOB error $e_{b}$
  * For each feature $j$, randomly permute its values in the OOB samples and calculate the error $e_{b,j}$
  * The importance of feature $j$ is the average increase in error:
  $$I(j) = \frac{1}{B}\sum_{b=1}^{B}(e_{b,j} - e_{b})$$
  * A large value indicates that permuting the feature's values increases the model error substantially, implying the feature is important

*6. Theoretical Properties*

* **Bias and Variance:**
  * Random Forests maintain the low bias of decision trees while reducing variance through ensemble averaging
  * The generalization error for Random Forests converges as the number of trees increases, without overfitting

* **Strength-Correlation Theorem:**
  * The generalization error of a Random Forest is bounded by:
  $$\text{Error} \leq \frac{\bar{\rho}(1 - s^2)}{s^2}$$
  where:
    * $\bar{\rho}$ is the average correlation between pairs of trees
    * $s$ is the strength of the trees (accuracy of individual trees)
  * Reducing correlation $\bar{\rho}$ (through feature randomization) and increasing strength $s$ (through good feature selection) improves performance
    
* **Law of Large Numbers:**
  * As the number of trees $B$ increases, the Random Forest prediction converges to:
  $$\lim_{B \to \infty} \frac{1}{B}\sum_{b=1}^{B}T_b(x) = E_{T}[T(x)]$$
  * This expected value over all possible trees prevents overfitting when adding more trees

*7. Hyperparameters*

* **Number of Trees ($B$):** Higher values provide better performance but increase computational cost. Typically ranges from 100 to 1000.
* **Maximum Features ($m$):** The number of features to consider at each split, typically $\sqrt{p}$ for classification and $p/3$ for regression.
* **Maximum Depth:** Controls the complexity of individual trees. Deeper trees capture more detail but may overfit.
* **Minimum Samples Split:** The minimum number of samples required to split a node.
* **Minimum Samples Leaf:** The minimum number of samples required to be at a leaf node.
* **Bootstrap:** Whether to use bootstrap samples (True) or the whole dataset (False) for building trees.
